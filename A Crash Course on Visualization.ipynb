```python
#Importing Libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Lets create a simple plot 
x = [0,1,2]
y = [100,200,300]
plt.plot(x,y)
```

```python
# If we want to clear description of plot 
plt.plot(x,y);
```

```python
# Dataframe creation
housing= pd.DataFrame({'rooms':[1,1,2,2,2,3,3,3],
                      'price':[100,120,190,200,230,300,330,305]
                      })
```

```python
housing 
```

```python
# Lets Create a Scatter Plot
plt.scatter(housing['rooms'],housing['price']);
```

```python
# By default line graph is created
plt.plot(housing['rooms'],housing['price']);
```

```python
# We can add different variables such as color and marker
plt.scatter(housing['rooms'],housing['price'], color = 'red', marker ='x');
plt.title('My scatter plot')
plt.xlabel('Room')
plt.ylabel('Price')
```

```python
# Now lets import seaborn and work with it
import seaborn as sns
```

```python
# We will use hearts dataset for our program
df = pd.read_csv('../DATA/heart.csv')
```

```python
df.head()
```

```python
# Lets use age for distribution plot
sns.distplot(df['age'])
```

```python
# You can play around with different variables such as color, bins and kde
plt.figure(figsize=(12,8))
sns.distplot(df['age'], kde=False, color = 'red', bins=25);
```

```python
df.head(3)
```

```python
# Count plot is an important plot to check the data imbalance
sns.countplot(x='sex',data=df)
```

```python
sns.countplot(x='target',data=df)
```

```python
sns.countplot(x='cp',data=df)
```

```python
# Box plot can help us to visualize data distribution
sns.boxplot(x='sex', y='age',data=df)
```

```python
# To visualize each varuable we can use pair plot
sns.pairplot(df)
```
